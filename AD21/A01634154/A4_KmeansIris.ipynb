```python
# Import the packages that we will be using
import matplotlib.pyplot as plt
import pandas as pd                 # For data handling

# Dataset url
url = "datasets/iris/iris.csv"

# Load the dataset
dataset = pd.read_csv(url, names=['Largosepalo','Anchosepalo','Largopetalo','Anchopetalo','Especie'] )
```

```python
dataset
```

```python
dataset = dataset.replace({"Especie":  {"Iris-setosa":0, "Iris-versicolor":1, "Iris-virginica":2} })

#Visualize the dataset
dataset
```

```python
plt.scatter(dataset.Largopetalo,dataset.Anchopetalo)
plt.title('Largo de pétalo vs Ancho de pétalo')
plt.xlabel('Largo de pétalo')
plt.ylabel('Ancho pétalo')
plt.show()
```

```python
df1 = dataset[dataset.Especie==0]
df2 = dataset[dataset.Especie==1]
df3 = dataset[dataset.Especie==2]

plt.scatter(df1.Largopetalo, df1.Anchopetalo, Label='Iris-setosa')
plt.scatter(df2.Largopetalo, df2.Anchopetalo, Label='Iris-versicolor')
plt.scatter(df3.Largopetalo, df3.Anchopetalo, Label='Iris-virginica')

plt.title('Largo de pétalo vs Ancho de pétalo')
plt.xlabel('Largo de pétalo')
plt.ylabel('Ancho de pétalo')
plt.legend()
plt.show()
```

```python
# Import sklearn KMeans
from sklearn.cluster import KMeans

# Define number of clusters
km = KMeans(n_clusters=3)

# Do K-means clustering (assing each point in the dataset to a cluster)
#yp = km.fit_predict(dataset)
yp = km.fit_predict(dataset[['Largopetalo','Anchopetalo']])

# Print estimated cluster of each point in the dataser
yp
```

```python
dataset['Cluster1'] = yp

dataset
```

```python
dataset.Cluster1.unique()
```

```python
# ...
km.cluster_centers_
```

```python
# ...
km.inertia_
```

```python
df1 = dataset[dataset.Cluster1==0]
df2 = dataset[dataset.Cluster1==1]
df3 = dataset[dataset.Cluster1==2]

plt.scatter(df1.Largopetalo, df1.Anchopetalo, Label='Iris-setosa')
plt.scatter(df2.Largopetalo, df2.Anchopetalo, Label='Iris-versicolor')
plt.scatter(df3.Largopetalo, df3.Anchopetalo, Label='Iris-virginica')

plt.scatter(km.cluster_centers_[:,0], km.cluster_centers_[:,1], color='magenta', marker='*', label='Centroides', s=256)

plt.title('Largo de pétalo vs Ancho de pétalo')
plt.xlabel('Largo pétalo')
plt.ylabel('Ancho pétalo')
plt.legend()
plt.show()
```

```python
#Caso 1

sse = []
k_rng = range(1,10)
for k in k_rng:
    km = KMeans(n_clusters=k)
    km.fit_predict(dataset[['Largopetalo','Anchopetalo']])
    sse.append(km.inertia_)
```

```python
plt.plot(k_rng,sse)

plt.title('Elbow plot')
plt.xlabel('K')
plt.ylabel('Sum of squared error')
plt.show()
```
